```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
file_path = "https://raw.githubusercontent.com/sanketkangle/Exploratory-Data-Analysis/master/Seaborn_tutorial/Pokemon.csv"
df = pd.read_csv(file_path, index_col = 0)
df.head()
```

```python
df.describe()
```

```python
sns.lineplot(x= "Defense", y="Attack", data=df)
```

```python
sns.lineplot(x= "Sp. Def", y="Sp. Atk", data=df)
plt.title("Trend in Special Attack and Special Defence")
plt.savefig("one.png")
```

```python
df.Legendary.value_counts()
```

```python
sns.lineplot(x= "Sp. Def", y="Sp. Atk", data=df, label="special")
sns.lineplot(x= "Defense", y="Attack", data=df, label="regular")
```

```python
# Count Plot (a.k.a. Bar Plot)
sns.countplot(x='Type 1', data=df)
 
# Rotate x-labels
plt.xticks(rotation=-45)

plt.savefig("one.png")
```

```python
sns.barplot(x="Type 1", y="Total", data=df)
plt.xticks(rotation=-45)
```

```python
corr = df.corr()
plt.figure(figsize=(10,6))
sns.heatmap(data=corr, annot=True)
plt.savefig("one.png")
```

```python
sns.scatterplot(x='Attack', y='Defense', data=df, hue='Stage')  
plt.savefig("one.png")
```

```python
sns.regplot(x='Attack', y='Defense', data=df)  
```

```python
sns.lmplot(x='Attack', y='Defense', data=df, hue="Stage")  
plt.savefig("one.png")
```

```python
plt.figure(figsize=(12,5))
sns.swarmplot(x='Type 1', y='Attack', data=df)
plt.savefig("one.png")
```

```python
sns.distplot(a=df['HP'], kde=False, bins=20)
plt.savefig("one.png")
```

```python
sns.distplot(a=df['Speed'], label="Speed",kde=False)
sns.distplot(a=df['HP'], label="HP", kde=False)
plt.legend()
plt.savefig("one.png")
```

```python
sns.distplot(a=df['HP'], kde=True, bins=20)
plt.savefig("one.png")
```

```python
sns.kdeplot(data=df.Total, shade=True )
plt.savefig("one.png")
```

```python
sns.kdeplot(data=df.Total, shade=True )
sns.kdeplot(data=df.HP, shade=True )
sns.kdeplot(data=df.Speed, shade=True )
plt.savefig("one.png")

```

```python
sns.jointplot(x="HP", y="Speed", data=df, kind="kde" )
plt.savefig("one.png")
```
